<h4>Using Python variables in the SQL statements </h4>

<span>To use a python variable in the SQL statement, add a ":" prefix to it.</span>
</br>
<span><i>For example</i>: we can create a dynamic SQL query statement retrieve result based on any country we want to pass, simply use <code>country</code> varibale in the following way.</span>

```python
# A python variable
country  = "Canada"

%sql select * from INTERNATIONAL_STUDENT_TEST_SCORES where country = :country
```

<h4>Store results from SQL eury into a python variable</h4>

```python
test_score_distribution = %sql SELECT test_score as "Test Score", count(*) as "Frequency" from INTERNATIONAL_STUDENT_TEST_SCORES GROUP BY test_score

# print the result
test_score_distribution
```

<h4>Use Pandas <code>Dataframe</code> to store the results</h4>

```python
import pandas as pd

pd_df= pd.DataFrame(test_score_distribution)

# check pd_df type
print("pd_df type = ", type(pd_df))

# print first 5 records
pd_df.head()
```

<span>We can use Pandas.DataFrame to store the results. It will be more convenient this way as it will make initial data analysis and data preprocessing</span>

```python
columns= ['Test Score', 'Frequency']

pd_df= pd.DataFrame(test_score_distribution, columns=columns)

# print first 5 records
pd_df.head()
```

<span>If you noticed, we got 0 and 1 as column names in the result.</span></br><span>We can manually mention columns while creating a DataFrame objecct</span>
